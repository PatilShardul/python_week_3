```python
import matplotlib.pyplot as plt

with open("test.txt") as file:
    file_data = file.read()
file_data = file_data.split('\n')

x = [row.split(' ')[0] for row in file_data]
y = [row.split(' ')[1] for row in file_data]

plt.title('Test File Graph')

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.plot(x, y,marker='o')
plt.show()
```
