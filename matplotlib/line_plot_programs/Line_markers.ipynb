```python
import matplotlib.pyplot as plt

player_x_axis=['Sachin','Sehwag','Dhoni','Yuvraj']
matches=[463,251,350,304]
runs=[18426,8273,10773,8701]

plt.style.use('ggplot')
plt.title('Indian player Stats')
plt.xlabel('Player Name')

plt.plot(player_x_axis,matches,color='b',linestyle='--',label='Matches played',marker='o')
plt.plot(player_x_axis,runs,color='r',linestyle='--',linewidth=3,label='Total Run score',marker='.')
plt.legend()
plt.tight_layout()

plt.show()
```
