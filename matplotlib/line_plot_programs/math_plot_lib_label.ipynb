```python
import matplotlib.pyplot as plt

x_axis=['Sachin','Sehwag','Dhoni','Yuvraj']
y_axis=[463,251,350,304]

plt.title('Indian player matches')
plt.xlabel('Player Name')
plt.ylabel('No. Of ODI Matches')

plt.plot(x_axis,y_axis,color='b',linestyle='--',label='Match Stat')
plt.legend()
plt.tight_layout()

plt.show()
```
