```python
import matplotlib.pyplot as plt

player_x_axis=['Sachin','Sehwag','Dhoni','Yuvraj']
matches_y_axis=[463,251,350,304]
runs_y_axis=[18426,8273,10773,8701]
plt.style.use('ggplot')
plt.title('Indian player Stats')
plt.xlabel('Player Name')

plt.plot(player_x_axis,matches_y_axis,color='b',linestyle='--',label='Matches played')
plt.plot(player_x_axis,runs_y_axis,color='r',linestyle='--',label='Total Run score')
plt.legend()
plt.tight_layout()

plt.show()
```
