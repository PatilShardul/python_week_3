# Write a Python programming to create a pie chart of gold medal achievements of five most successful countries in 2016 Summer Olympics. Read the data from a csv file.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
medal_data_frame= pd.read_csv('gold_medals_2016.csv')
plt.style.use('seaborn')
```

```python
country=medal_data_frame['country']
no_of_medals=medal_data_frame['gold_medal']
```

```python
plt.pie(no_of_medals,labels=country)
plt.title("gold medals  won by countries in 2016 olympics")
plt.show()
```
