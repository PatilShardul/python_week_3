# Write a Python program to draw a scatter plot using random distributions to generate balls of different sizes

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
size = [20,50,100,200,500,1000,60,90,10,300,600,800,75]
plt.scatter(x, y, s=size, color='none', edgecolors='black')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.tight_layout()
plt.show()
```
